```python
#plotting a subset of data
#set the width and the height of the figure
plt.figure(figsize=(12, 6))
#Add title
plt.title('Nobel Award Winners Year of Recieving Award')
sns.lineplot(data=project['id'], 
             label = 'ID Number')
#line chart showing year
sns.lineplot(data = project['year'], 
             label= 'Award Year')

#add label for horizontal axis
plt.xlabel('ID Number')
plt.ylabel('Award Year')
```

```python
#creat a bar chart shwoing avergae delay for NK ariline flights by month 
#set the size of the plot
plt.figure(figsize=(10, 5))

#set the title
plt.title('Nobel Award Winners Year of Recieving Award')

#the bar chart
sns.barplot(x= project.index, y=project.year)

#set the y axis label
plt.ylabel('Award Year')
plt.xlabel('Index')
```
